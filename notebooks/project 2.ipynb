```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Load the data
import io
df = pd.read_csv(io.BytesIO(uploaded['165+PRA+PAC_cleaned.csv']))

print(df.columns)

# Convert Gender to readable labels
df['Gender_Label'] = df['Gender'].map({1: 'Male', 2: 'Female'})

# Boxplot 1: PAC by Diagnosis (UPA vs BPA)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnosis(PA)', y='PAC', data=df)
plt.title('PAC Levels: UPA vs BPA')
plt.xlabel('Diagnosis (PA)')
plt.ylabel('Plasma Aldosterone Concentration (PAC)')
plt.grid(True)
plt.show()

# Boxplot 2: Age by Diagnosis
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnosis(PA)', y='Age', data=df)
plt.title('Age Distribution: UPA vs BPA')
plt.xlabel('Diagnosis (PA)')
plt.ylabel('Age')
plt.grid(True)
plt.show()

# Boxplot 3: Gender by Diagnosis
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnosis(PA)', y='Gender', data=df)
plt.title('Gender Code Distribution: UPA vs BPA')
plt.xlabel('Diagnosis (PA)')
plt.ylabel('Gender (1 = Male, 2 = Female)')
plt.grid(True)
plt.show()

#Age by Diagnosis split by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnosis(PA)', y='Age', hue='Gender_Label', data=df)
plt.title('Age Distribution by Diagnosis and Gender')
plt.xlabel('Diagnosis (PA)')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.grid(True)
plt.show()

```
